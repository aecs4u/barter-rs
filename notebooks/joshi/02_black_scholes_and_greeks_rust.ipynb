# Joshi Part 7: Black-Scholes & Greeks (Rust)

Closed-form pricing using `barter_joshi::black_scholes`.

**Kernel**: evcxr (Rust Jupyter kernel)

```rust
:dep barter-joshi = { path = "../../barter-joshi" }
```

## 1. Call and Put Prices

```rust
use barter_joshi::black_scholes::*;

let (s, k, r, vol, t) = (100.0, 100.0, 0.05, 0.2, 1.0);

let c = call_price(s, k, r, vol, t);
let p = put_price(s, k, r, vol, t);
let parity = put_call_parity_residual(s, k, r, vol, t);

println!("Call:  {c:.6}");
println!("Put:   {p:.6}");
println!("Put-call parity residual: {parity:.2e}");

println!("\nDigital Call: {:.6}", digital_call_price(s, k, r, vol, t));
println!("Digital Put:  {:.6}", digital_put_price(s, k, r, vol, t));
```

## 2. All Greeks

```rust
println!("{:<10} {:>12}", "Greek", "Value");
println!("{:-<24}", "");
println!("{:<10} {:>12.6}", "Delta", call_delta(s, k, r, vol, t));
println!("{:<10} {:>12.6}", "Gamma", gamma(s, k, r, vol, t));
println!("{:<10} {:>12.6}", "Vega", vega(s, k, r, vol, t));
println!("{:<10} {:>12.6}", "Theta", call_theta(s, k, r, vol, t));
println!("{:<10} {:>12.6}", "Rho", call_rho(s, k, r, vol, t));
```

## 3. Delta Across Spot Prices

```rust
println!("{:>8} {:>10} {:>12}", "Spot", "Delta", "Gamma");
println!("{:-<32}", "");
for spot_i in (60..=140).step_by(10) {
    let s = spot_i as f64;
    println!("{:>8.1} {:>10.4} {:>12.6}",
        s, call_delta(s, k, r, vol, t), gamma(s, k, r, vol, t));
}
```
