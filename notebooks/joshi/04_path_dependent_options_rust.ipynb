# Joshi Part 7: Path-Dependent Options (Rust)

Asian, barrier, and lookback options using `barter_joshi::path_dependent`.

**Kernel**: evcxr (Rust Jupyter kernel)

```rust
:dep barter-joshi = { path = "../../barter-joshi" }
:dep rand = "0.9"
```

## 1. Arithmetic Asian Call

```rust
use barter_joshi::path_dependent::*;

let mut rng = rand::rng();
let result = asian_call_arithmetic(100.0, 100.0, 0.05, 0.2, 1.0, 12, 100_000, &mut rng);

println!("Arithmetic Asian Call: {result}");
println!("(European vanilla ≈ 10.45 — Asian should be cheaper)");
```

## 2. Up-and-Out Barrier Call

```rust
println!("{:>10} {:>12} {:>12}", "Barrier", "Price", "Std Err");
println!("{:-<36}", "");

for barrier in [120.0, 130.0, 150.0, 200.0] {
    let mut rng = rand::rng();
    let r = barrier_up_and_out_call(100.0, 100.0, barrier, 0.05, 0.2, 1.0, 252, 100_000, &mut rng);
    println!("{:>10.0} {:>12.4} {:>12.4}", barrier, r.price, r.std_error);
}
println!("\nVanilla ≈ 10.45 (barrier → ∞)");
```

## 3. Down-and-Out Barrier Put

```rust
for barrier in [70.0, 80.0, 90.0] {
    let mut rng = rand::rng();
    let r = barrier_down_and_out_put(100.0, 100.0, barrier, 0.05, 0.2, 1.0, 252, 100_000, &mut rng);
    println!("Down-out Put (barrier={barrier:.0}): {r}");
}
```

## 4. Lookback Call

```rust
let mut rng = rand::rng();
let r = lookback_call(100.0, 0.05, 0.2, 1.0, 252, 100_000, &mut rng);
println!("Lookback Call: {r}");
println!("(Should be > Vanilla Call ≈ 10.45)");
```

## Exotic Option Price Comparison

| Option | Relation to Vanilla |
|--------|--------------------|
| Vanilla Call | Baseline ≈ 10.45 |
| Asian Call | Cheaper (averaging reduces volatility) |
| Barrier (up-and-out) | Cheaper (knock-out risk) |
| Lookback | More expensive (optimal strike) |
